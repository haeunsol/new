* 거래량 : 거래되는 자산의 양. 시장의 활기와 유동성 
* 거래가격 : 해당 거래에서 한 단위의 자산을 얻기 위해 지불하는 금액. 그 자산에 대한 시장 참여자들의 평가 반영
1. 상승 추세에서의 거래량과 가격 상관관계 : 일반적으로 상승 추세에서 거래량 늘어나는 경향. 높은 거래량 = 높아지는 관심 = 가격 상승 신호
2. 하락 추세에서의 거래량과 가격 상관관계 : 하락 추세에도 거래량이 늘어날 수 있지만 이는 판매 압력이 높아져 가격이 하락할 가능성이 있다는 신호
3. 종가 변동성 : 거래량 많은 경우 가격의 큰 변동 발생. 거래량이 낮은 경우 가격 안정화


```python
# 피어슨 상관계수 계산
correlation_coefficient = df['price'].corr(df['qty'])

print(f"price, qty 피어슨 상관계수: {correlation_coefficient}")
```

```python
correlation_coefficient = df['quote_qty'].corr(df['price'])

print(f"quote, price 피어슨 상관계수: {correlation_coefficient}")
```

```python
correlation_coefficient = df['quote_qty'].corr(df['qty'])

print(f"quote, qty 피어슨 상관계수: {correlation_coefficient}")
```
